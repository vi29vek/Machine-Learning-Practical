```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

# Based on our theory part, we will start the intercept from 0 and gradually we will see that we will reach 26.15

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color = 'red')
# best fit line
```

```python
# Lets apply gradient descent assuming slope is m=78.35 and it is constant
# we start b with 0
```

```python
y_pred = (78.35*x+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color = 'red', label = 'OLS')
plt.plot(x,y_pred,color = 'green', label = 'b=0')
plt.legend()
plt.show()
```

```python
epochs = 100
learning_rate = 0.01
m = 78.35
b = 0

for i in range(epochs):
    loss_slope = -2*np.sum(y-m*x.ravel()-b)
    b = b - (learning_rate*loss_slope)
    y_pred = m*x+b
    plt.plot(x,y_pred)
plt.scatter(x,y)
print(b)
```

```python
# Above we can see we reached the y_intercept in 100 epochs
```
